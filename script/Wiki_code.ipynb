```python
def detailed_ext(ss):
  uu=wikipedia.page(ss).url
  response = requests.get(url=uu)
  soup = BeautifulSoup(response.content, 'html.parser')
  allph=soup.find_all(['h1', 'h2', 'h3','p'])
  res=''
  n=len(allph)
  i=0
  while i<n:
    if str(allph[i].name) in ['h1', 'h2', 'h3']:
      asp=allph[i].text
      i+=1
      while i<n and str(allph[i].name)=='p':
        res+=(allph[i].text)
        i+=1
      #####################only find the first cell######################
      break 
    else:
      i+=1
  return res

def findkeys(pps,sw):
  bo=''
  for pp in pps:
    try:
      # print(wikipedia.summary(pp+sw))
      # bo+=(wikipedia.summary(pp+sw)) ## summary part
      ####### get all information before the next part (all paragraphs in introduction part)#########
      bo+=(detailed_ext(pp+sw)) 
    except:
      # print(pp)
      bo+=''  # Add none

  return bo

def wiki_extra(sample,sw):
  res=[]
  for i in range(len(sample)):
    pps= sample[i]
    # print(pps)
    res.append(findkeys(pps,sw))
  return res
  
```

```python
def text_processing(docs_orig):
    """
    input: list of text
    
    output: list of text, docs by text processsing
    """
    # words = set(nltk.corpus.words.words())
    # stops = set(stopwords.words('english'))
    # stopword_list = list(string.punctuation) + ['rt', 'via', '...','“', '”','’']
    # stops = stops.union(stopword_list)
#     docs_orig=df_sample['synopsis'].tolist()
    stops = set(stopwords.words('english'))
    stops.add('also')
    res=[]
    ################### clean data version1 ###########################################
    def clean(s):
      s=re.sub(r'[^\w\s]', ' ', s)  # removing punctuations
      # s=re.sub(r'[^A-Za-z]', ' ', s)
      words = set(nltk.corpus.words.words())# only keep alphas, only english(no other language), only letter(no digit)
      return " ".join(w for w in nltk.wordpunct_tokenize(s) if w.lower() in words and w.lower() not in stops) # removing the stopwords and make it lower case
    
    for i in range(len(docs_orig)):
      if docs_orig[i] is None:
          res.append("None") # None value processing
      else:
        ###clean
          res.append(clean(docs_orig[i]))
    return res

def sort_coo(coo_matrix):
    tuples = zip(coo_matrix.col, coo_matrix.data)
    return sorted(tuples, key=lambda x: (x[1], x[0]), reverse=True)

def extract_topn_from_vector(feature_names, sorted_items, topn=2):
    """get the feature names and tf-idf score of top n items"""
    
    #use only topn items from vector
    sorted_items = sorted_items[:topn]
    score_vals = []
    feature_vals = []
    
    # word index and corresponding tf-idf score
    for idx, score in sorted_items:
        #keep track of feature name and its corresponding score
        score_vals.append(round(score, 3))
        feature_vals.append(feature_names[idx])
    #create a tuples of feature,score
    #results = zip(feature_vals,score_vals)
    results= {}
    for idx in range(len(feature_vals)):
        results[feature_vals[idx]]=score_vals[idx]
    
    return results
def keyword_extraction(docs, cv, tfidf_transformer,topn=2):
    """
    input: text column from df
    
    output: list of keyword for each row of text
    """
    
#     docs = text_processing(df_dosc.tolist())
#     words = set(nltk.corpus.words.words())
#     stops = set(stopwords.words('english'))
    
#     cv=CountVectorizer(max_df=0.40, stop_words=stops, ngram_range = (1,4), max_features = 20000)
#     word_count_vector=cv.fit_transform(docs)
    
#     tfidf_transformer=TfidfTransformer(smooth_idf=True,use_idf=True)
#     tfidf_transformer.fit(word_count_vector)
    
#     vocabulary = cv.vocabulary_ #这些不能写在这个function里，vocabulary需要被单独save，类似hw4中用doc来存vocabulary
    
    keyword_list = []
    
    for doc in docs:
        feature_names=cv.get_feature_names()
        tf_idf_vector=tfidf_transformer.transform(cv.transform([doc]))

        sorted_items=sort_coo(tf_idf_vector.tocoo())

        keywords=extract_topn_from_vector(feature_names,sorted_items,topn)
        keylist = list(keywords.keys())
        
        keyword_list.append(keylist)
    
    return keyword_list
def wordAndVocabulary(df_dosc):
    def bigrams_per_line(doc):
      for ln in doc.split('\n'):
          terms = re.findall(r'\w{2,}', ln)
          for bigram in zip(terms, terms[1:]):
              yield '%s %s' % bigram
    docs = text_processing(df_dosc.tolist())
    words = set(nltk.corpus.words.words())
    stops = set(stopwords.words('english'))

    cv=CountVectorizer(max_df=0.40, stop_words=stopwords.words('english'), ngram_range = (1,3), max_features = 20000, analyzer=bigrams_per_line)
    word_count_vector=cv.fit_transform(docs)

    tfidf_transformer=TfidfTransformer(smooth_idf=True,use_idf=True)
    tfidf_transformer.fit(word_count_vector)

#     vocabulary = cv.vocabulary_ 
    return docs, cv, tfidf_transformer
```

```python
def take_keywords(col):
  df_docs, train_cv, tfidf_transformer = wordAndVocabulary(col)
  return keyword_extraction(df_docs, train_cv, tfidf_transformer,5)
```
